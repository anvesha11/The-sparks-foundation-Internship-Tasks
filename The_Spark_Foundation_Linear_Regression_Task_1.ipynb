```python
import pandas as pd
dataset = pd.read_csv('Hours_Scores.csv')
```

```python
#X means take all rows and all columns except last one (Hours)
X = dataset.iloc[:, :-1].values
#y means means take all rows and only columns
#with index 1 — In python indexes begin at 0 — so index 1 here is the second column (Scores)
y = dataset.iloc[:, 1].values
```

```python
#Step 2: Split data into training and testing
#test_size means 1/5 is used for testing and 4/5 for training
#random state is for comparing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/5, random_state=0)
```

```python
#Step 3: Fit Simple linear Regression Model to training data
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Step 4: Make prediction
y_pred = regressor.predict(X_test)
```

```python
# Step 5 - Visualize training set results
import matplotlib.pyplot as plt
#plot the actual points of training data set
plt.scatter(X_train, y_train, color = 'red')
#plot the regression line
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Hours spend studying VS the result obtained (Training set)')
plt.xlabel('Hours prent studying everyday')
plt.ylabel('Result')
plt.show()
```

```python
#Step 6: Visualising test set data
import matplotlib.pyplot as plt
#plot the actual data points of the test set
plt.scatter(X_test, y_test, color='red')
#plot the regression line (same as above)
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Hours spend studying VS the result obtained (Training set)')
plt.xlabel('Hours prent studying everyday')
plt.ylabel('Result')
plt.show()
```

```python
# Step 7 - Make new prediction
result_pred = regressor.predict([[9.25]])
```

```python
print('The predicted result/score of a student who studies for 9.25 hrs a day is ', result_pred)
```
